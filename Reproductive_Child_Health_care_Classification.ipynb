```python
y= rep_child_healthcare['fetal_health']
x=rep_child_healthcare.drop(labels=['fetal_health'],axis=1)
```

```python
corr= rep_child_healthcare.corr()
```

```python
plt.figure(figsize = (16,5))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, linewidths=0.5)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
print(y_train.shape)
print(y_test.shape)
print(x_test.shape)
print(x_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_train= sc.fit_transform(x_train)
x_test= sc.transform(x_test)
```

```python
from sklearn.decomposition import PCA
pca= PCA()
pca.fit(x_train)
```

```python
pca.n_components_
```

```python
pca.transform(x_train)
pca.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression(random_state=0)
reg.fit(x_train, y_train)
```

```python
pred=reg.predict(x_test)
pred
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, pred)
```

```python
prob=reg.predict_proba(x_test)
prob[0,:]
```

```python
from sklearn.metrics import mean_squared_error
error= mean_squared_error(y_test, pred)
np.sqrt(error)
```

```python
from sklearn.metrics import classification_report
cr= classification_report(y_test, pred)
print(cr)
```
